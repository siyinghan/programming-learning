### 从Oracle SQL Developer复制出来的数据添加表头

`column_name.csv`：表头文件

```python
# 提取表头名到list
header = pd.read_csv("column_name.csv", header=None)
header.columns = ["COLUMN_NAME"]
lst = header["COLUMN_NAME"].tolist()

# 把表头名更新到所需表
df = pd.read_csv("content.csv", header=None, delimiter='\t')
df.columns = lst
df.to_csv("T_BFG_009_ACTIVE.csv")
```
